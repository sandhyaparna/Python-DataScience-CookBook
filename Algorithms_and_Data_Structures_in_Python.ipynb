### Linked Lists
* Linked lists are composed of nodes and references/pointers pointing from one node to the other. The last reference is pointing to a null
* It is a collection of single data elements that are connected via references. Usually, each element of the linked list has the same data type that is specific to the list.
* A single list element is called a node. The nodes are not like arrays which are stored sequentially in memory. Instead, it is likely to find them at different memory segments, which you can find by following the pointers from one node to the next. It is common to mark the end of the list with a NIL element, represented by the Python equivalent None.
* No indexes

https://stackabuse.com/python-linked-lists/

* Linked lists have sequential access
* Each node is composed of a data and a reference/link to the next node in the sequence
* Single linked list: A node in a single-linked list only points to the next element in the list
![](https://s3.amazonaws.com/stackabuse/media/python-linked-lists-1.png)

* Double Linked list: A node in a double-linked list points to the previous node, too
![](https://s3.amazonaws.com/stackabuse/media/python-linked-lists-2.png)

* Adding and deleting of nodes as well as splitting single-linked lists is done in not more than two steps. In a double-linked list four pointers have to be changed.


Advantages:
* Linked lists are dynamic data structures (arrays are not)
* Very efficient if we want to maniplate the first elements. O(1) time complexity
* Can store iems with different sizes: an array assumes every element to be exactly the same
* It's easier for a linked list to grow organizally. An array's size needs to be know a head of time or re-created when it needs to grow

Insertion Operations:
https://healthedge.udemy.com/course/algorithms-and-data-structures-in-python/learn/lecture/4660390#notes
* Inserting items at the beginning of the linked list: we just have to update the references. 
* linkedlist.insertaAtStart(4): value 4 is inserted at the beginning
* Inserting items at the end the linked list is NOT that simple as we have to traverse the whole linked list to find the last node. The last node is the one pointing to a NULL. And we have to update the references when we get there. we are iterating through all the items here and so the time complexity is linear i.e O(n)
* linkedlist.insertAtEnd(25): Updating the references again at the end takes O(1) time complexity but we have to traverse the list itself and thats what take O(N)
* similarly removing items at the beginning has O(1) time complexity whereas removing items at a given point has O(n) time complexity
* Removing items for single LinkedList has O(N) time complexity as to remove a node we need node+ predecessor wheras double LinkedList has O(1) time complexity as we need only the node

* Difference between arrays and linked lists
  * Search operation is better in arrays O(1) than in linked lists as Arrays maintain index based system which makes it faster for seraching an element in the list. On the other hand LinkedList requires the traversal through all the items O(n) for seraching an element
  * Deletion operation is better in LinkedList as removal only requires change in the pointer location which can be done very fast.
  * In memory management, arrays are better as they don't need any extra memory but LinkedLists do need extra memory because of the reference/pointers

### Stacks
Operations: push(), pop() and peek()

*  A stack is a linear data structure that stores items in a Last-In/First-Out (LIFO i.e the last item we insert is the first item we take out) or First-In/Last-Out (FILO) manner. In stack, a new element is added at one end and an element is removed from that end only. can be visualized as a bucket of items
* push operation: Adding items to the stack. O(1)
* pop operation: Taking out the last item we inserted. O(1)
* peek operation: Return/show the item feom the top of the stack without removing it. O(1)

Stack memory | Heap memory
--- | ---
limited in size | no size limits
fast access | slow access
local variables | objects
space is managed efficiently by CPU | memory may be fragmented
variables cannot be resized | variables can be resized

* Recursion method involves stack. Recursive method calls are going to be piled up in the stack

Applications
* Back button in web browsers
* Undo operation
* stack memory stores local variables and function calls





```
class Stack:

	def __init__(self):
		self.stack = [] #1-D array
		
	def isEmpty(self):
		return self.stack == [] #checking if self.stack is equal to an empty array
  
  # O(1)	
	def push(self, data):
		self.stack.append(data) # Insert at end
	
  # O(1)
	def pop(self):
		data = self.stack[-1]
		del self.stack[-1]
		return data

  # O(1)	
	def peek(self): 
		return self.stack[-1] # Returns the last item of the 1-D array

	def sizeStack(self):
		return len(self.stack)
		
stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
print(stack.sizeStack())
print("Popped: ", stack.pop())
print("Popped: ", stack.pop()) # if we use stack.pop() when the array is empty, it gives error
print(stack.sizeStack())
print("Peek:", stack.peek())
print(stack.sizeStack())
print(stack.isEmpty())
```

### Queue
Operations: enqueue(), dequeue() and peek()
* FIFO structure: First In First Out
* Can be visualized as a funnel. first thing that enters from top is the one to get out from the bottom of the funnel
* Enqueue operation: adding new item to the end of the queue is simple
* Dequeue operation: Removing the item starting at the beginning of the queue
* 

Applications:
* When a resource is shared with several customers, we store them in a queue
* Operational research applications or stochastic models relies heavily on queues





```
#FIFO - first -> first

class Queue:

	def __init__(self):
		self.queue = []
		
	def isEmpty(self):
		return self.queue == []
		
	def enqueue(self, data):
		self.queue.append(data) # Adding items at the end of the array
		
	def dequeue(self):
		data = self.queue[0] # Extracting first item in the array as it is the first to enter into the array
		del self.queue[0]
		return data
		
	def peek(self):
		return self.queue[0]
		
	def sizeQueue(self):
		return len(self.queue)
	
queue = Queue()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print(queue.sizeQueue())
print("Dequeue: ", queue.dequeue())
print("Dequeue: ", queue.dequeue())
print(queue.sizeQueue())
```

### Binary Search Trees
* O(log N) time  complexity
* Root node --> Parent node --> Leaf node
* In Binary search trees every node can have at most two children: left and right. Left child value is smaller than the parent and right child value is greater than the parent
* It is good because we get rid of half of the data in which we are seraching
* Tree should be balanced and height is log N
* Array representation of a binary search tree









* Insertion Operation: We start at the root node. If the data we want to insert is greater than the root node we go to the right, if it is smaller, we go to the left and so on. A value less than the parent node is always on the left side only
* Search Operation: If the data we want to find is greater than the root node we go to the right, if it is smaller, we go to the left until we find it. 
  * To find the smallest value of the binary tree we need to go to left as far as possible as that is where the smallest value will be
  * To find the largest value of the binary tree we need to go to right as far as possible as that is where the largest value will be
* Delete Operation: We don't remove the node from the tree we just mark that it has been removed
  * when the node we want to get rid of is a leaft node, set it to null 
  * when the node we want to get rid of has a single child, we just have to remove it and update the parent nodes reference to point it to the leaf node
  * when the node we want to get rid of has two children, find the largest value (predecessor) in the left subtree or the smallest item (successor) in the right sub tree
    * We swap the node we want to remove with the predecessor
    * We can swap with the successor also but needs to update references of the child nodes
* Traversal: Sometimes it is necessary to visit every node in the tree. we can do it several ways:
  * In-order traversal: we visit the left subtree + Parent node + right sub tree recursively
  * Pre-order traversal: We visit Root node left subtree + right sub tree recursively 
  * Post-order traversal: We visit left sub tree + right sub tree + the root recursively
![](https://leetcode.com/articles/Figures/145_transverse.png)
![](https://www.cdn.geeksforgeeks.org/wp-content/uploads/Preorder-from-Inorder-and-Postorder-traversals.jpg)

* Binary search trees can be unbalanced trees when we are constructing from a sorted array and this causes the running time to inc to O(N) from O(log N). For eg: if we want to insert 1,2,3,4 in the mentioned order, all the items will be kept joining as the right node

Applications:
* Operating system file systems
* ML: Decision trees and boosting uses very simple tree structures

```
class Node:

    def __init__(self, data, parent):
        self.data = data
        self.leftChild = None
        self.rightChild = None
        self.parent = parent


class BinarySearchTree:

    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data, None)
        else:
            self.insert_node(data, self.root)

    # O(logN) BUT if the tree is balances (left subtree contains app. same amount of item then right subtree)
    def insert_node(self, data, node):

        # we have to go to the left subtree recursively
        if data < node.data: # If data is less than node
            if node.leftChild: # if left child is not empty
                self.insert_node(data, node.leftChild) # repeat the above procedure
            else: # if left child is empty
                node.leftChild = Node(data, node) # insert data as left child
        # we have to visit the right subtree
        else: # If data is greater than or equal to the node
            if node.rightChild:
                self.insert_node(data, node.rightChild)
            else:
                node.rightChild = Node(data, node)

    def remove_node(self, data, node):

        if node is None:
            return

        if data < node.data:
            self.remove_node(data, node.leftChild)
        elif data > node.data:
            self.remove_node(data, node.rightChild)
        else:

            if node.leftChild is None and node.rightChild is None:
                print("Removing a leaf node...%d" % node.data)

                parent = node.parent

                if parent is not None and parent.leftChild == node:
                    parent.leftChild = None
                if parent is not None and parent.rightChild == node:
                    parent.rightChild = None

                if parent is None:
                    self.root = None

                del node

            elif node.leftChild is None and node.rightChild is not None:  # node !!!
                print("Removing a node with single right child...")

                parent = node.parent

                if parent is not None:
                    if parent.leftChild == node:
                        parent.leftChild = node.rightChild
                    if parent.rightChild == node:
                        parent.rightChild = node.rightChild
                else:
                    self.root = node.rightChild

                node.rightChild.parent = parent
                del node

            elif node.rightChild is None and node.leftChild is not None:
                print("Removing a node with single left child...")

                parent = node.parent

                if parent is not None:
                    if parent.leftChild == node:
                        parent.leftChild = node.leftChild
                    if parent.rightChild == node:
                        parent.rightChild = node.leftChild
                else:
                    self.root = node.leftChild

                node.leftChild.parent = parent
                del node

            else:
                print('Removing node with two children....')

                predecessor = self.get_predecessor(node.leftChild)

                temp = predecessor.data
                predecessor.data = node.data
                node.data = temp

                self.remove_node(data, predecessor)

    def get_predecessor(self, node):
        if node.rightChild:
            return self.get_predecessor(node.rightChild)

        return node

    def remove(self, data):
        if self.root is not None:
            self.remove_node(data, self.root)

    def traverse(self):
        if self.root is not None:
            self.traverse_in_order(self.root)

    def get_max_value(self):
        if self.root:
            return self.get_max(self.root)

    def get_max(self, node): # max node is the right most node that doen't have a child

        actual = self.root

        while actual.rightChild is not None:
            actual = actual.rightChild

        return actual.data

    def get_min_value(self): # min node is the left most node that doen't have a child

        if self.root:
            return self.get_min(self.root)

    def get_min(self, node):
        if node.leftChild:
            return self.get_min(node.leftChild)

        return node.data

    def traverse_in_order(self, node):

        if node.leftChild:
            self.traverse_in_order(node.leftChild)

        print('%s' % node.data)

        if node.rightChild:
            self.traverse_in_order(node.rightChild)


bst = BinarySearchTree()
bst.insert(10)
bst.insert(5)
bst.insert(-5)
bst.insert(1)
bst.insert(99)
bst.insert(34)
bst.insert(1000)

bst.remove(99)

bst.traverse()
```

### Balanced Search Trees - Red-Black trees
* Each node is either red or black
* Root node and NIL/NULL are always black
* Every red node must have 2 black child nodes and no other children i.e it must have a black parent
* Every path from a given node to any of its descendant NIL/NULL nodes contains the same number of black nodes
* Every new node is red by default
* Inserting is the same way as we have seen for binary search trees or AVL trees
* On every insertion we have to check whether we have violated the red-black properties or not
* If we have violated the RB properties: we have to rebalance the tree:
  * make rotations
  * just recolor the nodes
* It is not rigidly balanced aslLongest possible path consists of 2*m nodes. No path is more than wice the same number of black nodes
![](https://www.includehelp.com/data-structure-tutorial/images/red-black-tree-1.jpg)


Difference between Red-black and AVL trees
* Linux kernels relies heavily on red-black trees
* Insertion is fast as it is not rigidly balanced
* For insertion/deletion intensive tasks, use Red-black trees
* For a look-up (search) intensive task use an AVL tree

### Priority Queue
* Every item has an additional property called priority value
* In priority queue, an element with high priority is served before an element with lower priority
* Priority queues are usually implemented with heaps, but it can be implemented with self balancing trees as well
* Very similar to queues with some modification: when we would like o get to the next item the highest priority element is retrieved first. No FIFO structure



* Insertion: We have to specify the data we want to insert and the priority associated with the given data
* Highest priority element: Returns the element with highest priority and we have to reconstruct the heap. Max heap & Min heap
* peek: Return the element with highest priority but the structure of the heap doesn't change

##### Heap
* It is basically a binary tree. Data is inerted line by line from left to right
* Heap is a special tree structure in which each parent node is less than or equal to its child node. Then it is called a Min Heap. The lowest key is the root node 
* If each parent node is greater than or equal to its child node then it is called a max heap. The highest key is in the root node
* It is very useful is implementing priority queues where the queue item with higher weightage is given more priority in processing. 
![](https://www.codesdope.com/staticroot/images/ds/binary_tree9.png)


Applications:
* Dijkstra's algo, Prims algo

Heap as array:
![](https://www.algolist.net/img/binary-heap-array-mapping.png)
* Insertion: For max heap, after every item insertion we need to swap items such that parent node is greater than child node. O(N) time complexity
* Inserting an item to the heap is just adding the data to the array with increment index
* Removal: 
  * When the node we want to remove is root node. Delete the root node, then a void will be created there so fill it with the last indexed node and then reconstruct tree by swapping such that parent node is always greater than child nodes. Delete of root node is O(1) and reconstructing tree is O(log N)
  * When the node we want to remove is arbitary node, we search for the item in entire array. O(N)

Heap sort: https://healthedge.udemy.com/course/algorithms-and-data-structures-in-python/learn/lecture/4688052#questions

* Use heap data structure rather than a linear-time search to find the max

In max heap we swap the root node and consider the highest value is sorted and then it is considered not to be part of the tree and check whether the remaining tree is a valid max heap or not









```

class Heap(object):

	HEAP_SIZE = 10
	
	def __init__(self):
		self.heap = [0]*Heap.HEAP_SIZE;
		self.currentPosition = -1;
		
	def insert(self, item):
	
		if self.isFull():
			print("Heap is full..");
			return 
			
		self.currentPosition = self.currentPosition + 1
		self.heap[self.currentPosition] = item
		self.fixUp(self.currentPosition)
		
	def fixUp(self, index):
	
		parentIndex = int((index-1)/2)
		
		while parentIndex >= 0 and self.heap[parentIndex] < self.heap[index]:
			temp = self.heap[index]
			self.heap[index] = self.heap[parentIndex]
			self.heap[parentIndex] = temp
			index=parentIndex
			parentIndex = (int)((index-1)/2)
			
	def heapsort(self):
	
		for i in range(0,self.currentPosition+1):
			temp = self.heap[0]
			print("%d " % temp)
			self.heap[0] = self.heap[self.currentPosition-i]
			self.heap[self.currentPosition-i] = temp
			self.fixDown(0,self.currentPosition-i-1)
			
	def fixDown(self, index, upto):

		while index <= upto:
		
			leftChild = 2*index+1
			rightChild = 2*index+2
	
			if leftChild < upto:
				childToSwap = None
				
				if rightChild > upto:
					childToSwap = leftChild
				else:
					if self.heap[leftChild] > self.heap[rightChild]:
						childToSwap = leftChild
					else:
						childToSwap = rightChild
				
				if self.heap[index] < self.heap[childToSwap]:
					temp = self.heap[index]
					self.heap[index] = self.heap[childToSwap]
					self.heap[childToSwap] = temp
				else:
					break
				
				index = childToSwap
			else:
				break;							
			
	def isFull(self):
		if self.currentPosition == Heap.HEAP_SIZE:
			return True
		else:
			return False
			
			
if __name__ == "__main__":

	heap = Heap()
	heap.insert(10)
	heap.insert(-20)
	heap.insert(0)
	heap.insert(2)
	heap.insert(4)
	heap.insert(5)
	heap.insert(6)
	heap.insert(7)
	heap.insert(20)
	heap.insert(15)
	
	heap.heapsort()
```

##### Binomial Heap
* It is a set of binomial trees where each binomial tree follows Min Heap property
* Similar to Binary heap but also supports quick merging of two heaps
* Insertion: O(log N) can be reduced to O(1) with help of binomial heaps
* A Binomial Tree of order 0 has 1 node. A Binomial Tree of order k can be constructed by taking two binomial trees of order k-1 and making one as leftmost child or other. 
A Binomial Tree of order k has following properties. 
  * It has exactly $2^k$ nodes. 
  * It has depth as k. 
  * There are exactly ${k \choose i}$ nodes at depth i for i = 0, 1, . . . , k. 
  * The root has degree k and children of root are themselves Binomial Trees with order k-1, k-2,.. 0 from left to right.

### Hashtables / Dictionaries
* Contains key-value pairs
* Insertion operation: O(1) but we must transform the keys into array indexes
* From Key space keys are maped to a slot in the array. Hashing maps a certain key of any type to a random array index
* Collision is when we map two or more keys to the same bucket. If hash function is perfect then no collisions at all
  * Collision can be resolved using chaining, we use linkedlist to store both the values at the same bucket. chaining is not a good solution as time complexity will be same a linkedlist
  * Other method is open addressing, we generate a new index for the item, new item is filled in the next available index/slot. Empty slot can be found by
    * Linear probing: keep checking for empty slot one index after other
    * Quadratic probing: We try 1,2,4,8 slots
    * Rehashing: we hash the result again in order to find an empty slot
* Load factor is number of keys divided by size of the array = $\frac{n}{m}$
  * If Load factor is 0 $\implies$ hashtable is empty
  * If Load factor is 1 $\implies$ hashtable is full
  * If Load factor is ~0 $\implies$ hashtable is nearly empty and alot of memory is wasted
  * If Load factor is ~1 $\implies$ hashtable is nearly full and performance will decrease and operations will be slow as probability of collisions are very very high. 
* Dynamic resizing is needed to reduce time complexity
  * Performance depends on load factor
  * Resize the table, when load factor exceeds the given threshold. In python threshold is $\frac{2}{3}$ ~ 0.66 $\implies$ when 66% of array is full then we rize underlying array data structure and resizing takes O(n) 

Applications:
* substring search
* Counting given word occuerence in a particular documen
* Lookup tables
* Cryptographic fingerprints
* Storing passwords: Hashes of the password are stored instead of the actual password

  












```
class HashTable(object):

	def __init__(self):
		self.size = 10
		self.keys = [None] * self.size
		self.values = [None] * self.size
			
	def put(self, key, data):
		
		index = self.hashfunction(key)
		
		# not None -> it is a collision !!!
		while self.keys[index] is not None:	
			if self.keys[index] == key:
				self.values[index] = data # update
				return
				
			# rehash try to find another slot
			index = (index+1)%self.size

		# insert
		self.keys[index] = key
		self.values[index] = data
	
	def get(self,key):
		
		index = self.hashfunction(key)
		
		while self.keys[index] is not None:
			if self.keys[index] == key:
				return self.values[index]
				
			index = (index+1)%self.size
		
		# it means the key is not present in the associative array
		return None

	def hashfunction(self, key): # Transforming key into array indices
		sum = 0
		for pos in range(len(key)):
			sum = sum + ord(key[pos]) # Get ASCII value

		return sum%self.size #

if __name__ == "__main__":

	table =HashTable()
	
	table.put("apple",10)
	table.put("orange",20)
	table.put("car",30)
	table.put("table",40)
	
	print(table.get("car"))
```

### Ternary Search Trees (Tries)
* Sorting is very very efficient
* It stores less references and null objects
* It stores characters or strings in nodes
* Each node has <= 3 children. left is smaller than the parent, middle is next character in a given word, Greater than the parent node
* It is faster than hashmap and more flexible than binary search trees
* Difference between Tries and TST is that for Tries root node have as many children as the letters in English alphabets i.e 26
* Combine Tries and TST such that at the root it is a trie with many many children and at lower levels it becomes a TST with only 3 children

Applications:
* Auto-complete spell-checker
* Near-neighbor searching
* prefix matching in google search


* Insert operation: 
  * If the character is smaller alphabetically, we go to the left
  * If the character is equal, we go to the middle
  * If the character is greater alphabetically, we go to the right

Inserting keywords: {'sign', 'sinus', 'sit', 'silly', 'side', 'son', 'soda', 'sauce', 'sand', 'soap', 'sar'}

* Extract value associated with a given key


```
class Node(object):

	def __init__(self, character):
		self.character = character
		self.leftNode = None
		self.middleNode = None
		self.rightNode = None
		self.value = 0
		
class TST(object):

	def __init__(self):
		self.rootNode = None
		
	def put(self,key,value):
		self.rootNode = self.putItem(self.rootNode, key, value, 0)
		
	def putItem(self, node, key, value, index):
	
		c = key[index]
		
		if node == None:
			node = Node(c)
			
		if c < node.character:
			node.leftNode = self.putItem(node.leftNode, key, value, index)
		elif c > node.character:
			node.rightNode = self.putItem(node.rightNode, key, value, index)
		elif index < len(key) - 1:
			node.middleNode = self.putItem(node.middleNode, key, value, index+1)
		else:
			node.value = value
			
		return node;
		
	def get(self, key):
	
		node = self.getItem(self.rootNode, key, 0)
		
		if node == None:
			return -1
			
		return node.value
	
	def getItem(self, node, key, index):
	
		if node == None:
			return None
		
		c = key[index]
		
		if c < node.character:
			return self.getItem(node.leftNode, key, index)
		elif c > node.character:
			return self.getItem(node.rightNode, key, index)
		elif index < len(key) - 1:
			return self.getItem(node.middleNode, key, index+1)
		else:
			return node
			
if __name__ == "__main__":

	tst = TST()
	
	tst.put("apple",100)
	tst.put("orange",200)
	
	print( tst.get("orange") )
```

#### Selection sort

#### Insertion sort
Iterate from array[1] to array[n] over the array. Compare the current key to its predecessor. If the ke element is smaller than its predecessor, compare it to the elements before. Move the greater elements one position up to make space for the swapped element

```
nums = [3,2,1,5,4]
def swap(nums, i, j):
	temp = nums[i]
	nums[i] = nums[j]
	nums[j] = temp

for i in range(len(nums)):
  
  j = i #initialize j for using in while loop
  
  while j>0 and nums[j-1] > nums[j]: # every item that preceeds nums[j]
    swap(nums,j,j-1)
    j = j - 1

  print(nums) 
```

#### Quick Sort
* Used to sort primitive types
* Outperforms heapsort and merge sort. It is a divide & conquer algo. 
* Divide the collection in two roughly equal parts by taking a pseudo random element and using it as a pivot.(choose middle index) 
* Elements smaller than the pivot get moved to the left of the pivot and elements larger than the pivot to the right of it. This process is repeated for the collection to the left of the pivot as well as for the array of elements to the right of the pivot until the whole array is sorted

```
def quick_sort(nums,low,high):
    
	if low >= high:
		return
		
	pivot_index = partition(nums,low,high)
	quick_sort(nums,low, pivot_index-1)
	quick_sort(nums, pivot_index+1,high)
	
	
def partition(nums,low,high):

	pivot_index = (low+high)//2
	swap(nums,pivot_index,high)
	
	i = low
	
	for j in range(low,high,1):
		if nums[j] <= nums[high]:
			swap(nums,i,j)
			i = i + 1
			
	swap(nums,i,high)
	
	return i
	
def swap(nums,i,j):
	temp = nums[i]
	nums[i] = nums[j]
	nums[j] = temp
	
	
if __name__ == "__main__":
   
   nums = [-2,-1,0,1,0,-1,-2]
   quick_sort(nums,0,len(nums)-1)
   print(nums)
```

#### Merge Sort
* Used to sort reference types/objects
* Comparision based algo
* MergeSort(arr[], l,  r)
  * If r > l
      1. Find the middle point to divide the array into two halves:  
              middle m = (l+r)/2
      2. Call mergeSort for first half:   
              Call mergeSort(arr, l, m)
      3. Call mergeSort for second half:
              Call mergeSort(arr, m+1, r)
      4. If there is only a single item in the sub array we consider it to be sorted by definition
      5. Merge the two halves sorted in step 2 and 3:
              Call merge(arr, l, m, r)

![](https://www.101computing.net/wp/wp-content/uploads/Merge-Sort-Algorithm-900x506.png)

```
def merge_sort(nums):
    
	if len(nums) == 1:                                  
		return
		
	middle_index = len(nums) // 2
		
	left_half = nums[:middle_index]
	right_half = nums[middle_index:]
	
	merge_sort(left_half)
	merge_sort(right_half)
	
	i = 0
	j = 0
	k = 0
	
	while i<len(left_half) and j<len(right_half):
		if left_half[i] < right_half[j]:
			nums[k] = left_half[i]
			i = i + 1
		else:
			nums[k] = right_half[j]
			j = j + 1
			
		k = k + 1
		
	while i<len(left_half):
		nums[k] = left_half[i]
		k = k + 1
		i = i + 1		
	
if __name__ == "__main__":
   
   nums = [-3,-2,-1,1,2,1,0,-1,-2,-3]
   merge_sort(nums)
   print(nums)
```

#### Hybrid Algos: IntroSort, TimSort
* It combines more algos to solve a given problem, choses one algo depending on the data or switching between them over the course of the algo
* IntroSort = QuickSort + HeapSort
* TimSort = INsertionSort + MergeSort. Best case O(N), worst case O(N log N)

#### Non-Comparision based sorting
* Couting sort: Sorts integers only. O(K+N)
  * Find out the maximum element (let it be max) from the given array.
  * Initialize an array of length max+1 with all elements 0. This array is used for storing the count of the elements in the array.
  * Store the count of each element at their respective index in count array
