```
# specifying which ones are to be considered as missing values
missing_value_formats = ["n.a.","?","NA","n/a", "na", "--"]
Employees = pd.read_csv("employees.csv", na_values = missing_value_formats)
```

### Histogram

```
fig = px.histogram(Employees, x="Salary")
fig.show()
```

### Q-Q plots

```
from scipy import stats

qq = stats.probplot(Employees['Salary'], sparams=(1))
x = np.array([qq[0][0][0], qq[0][0][-1]])

fig = go.Figure()
fig.add_scatter(x=qq[0][0], y=qq[0][1], mode='markers')
fig.add_scatter(x=x, y=qq[1][1] + qq[1][0]*x, mode='lines')
fig.layout.update(showlegend=False)
fig.show()
```

### Box plots

```
import plotly.express as px
fig = px.box(Employees, y="Salary")
fig.show()
```
